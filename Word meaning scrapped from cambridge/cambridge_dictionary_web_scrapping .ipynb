```python
files = ["word1.csv", "meaning1.csv", "sentences1.csv"]
csv_files=[]
for file in files:
    csv_files.append(open(file, 'w'))
writers=[]
for csv_file in csv_files:
    writer=csv.writer(csv_file)
    writers.append(writer)

header = [['wordid', 'word'],['wordid', 'meaningid', 'meaning'],['meaningid', 'sentenceid', 'examples']]
headercount=0
for writer in writers:
    writer.writerow(header[headercount])
    headercount+=1
for csv_file in csv_files:
    csv_file.close()
```

```python
wordId = 1
meaningId = 1
sentenceId = 1
for word in words:
    url = 'https://dictionary.cambridge.org/dictionary/english/'+ word
    while True:
        try:
            html_text = requests.get(url, headers = headers).text
            break
        except:
            print("Connection refused by the server..")
            print("Let me sleep for 5 seconds")
            print("ZZzzzz...")
            time.sleep(1)
            print("Was a nice sleep, now let me continue...")
        
    soup = BeautifulSoup(html_text, 'lxml')
    blocks = soup.find_all('div', class_='def-block ddef_block', limit = 2)
    for block in blocks:
        meaning = block.find('div', class_='def ddef_d db').text[:-2]
        sen = block.find_all( 'span', class_= 'eg deg', limit = 2)
        for s in sen:
            sentence = s.text
            row3 = [meaningId,sentenceId, sentence]
            writer3.writerow(row3)
            sentenceId += 1
        row2 = [wordId,meaningId,meaning]
        writer2.writerow(row2)
        meaningId += 1
    row1 = [wordId,word]
    writer1.writerow(row1)
    wordId += 1
    print(f'{wordId} word completed')
csv_file.close()
```
